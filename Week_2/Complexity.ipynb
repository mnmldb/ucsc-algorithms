### Time: $O(n)$ Space: $\Theta(1)$ 

```python
def findmax(l: 'list') -> 'int':
    largest = l[0]
    n = len(l)
    for i in range(1,n):
        if l[i] > largest:
            largest = l[i]
    return largest
```

```python
def squareroot(n:int) -> 'int':
    i = 2
    while (i * i <= n):
        i = i + 1
    return i-1
print(squareroot(100))
print(squareroot(29))
print(squareroot(1000))
```

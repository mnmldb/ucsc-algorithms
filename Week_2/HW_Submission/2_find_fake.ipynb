```python
def find_fake(l:'list', f:'int')->"list of size 2":
    #a[0] = pos
    #a[1] = num_steps
    a = [0,0]
    n = len(l)
    for i in range(0,n):
        if (l[i] == 1):
            a[0] = i
            a[1] = i
    return a
```

```python
def find_fake_smart(l:'list', f:'int')->"list of size 2":
    a = [0, 0] # output
    n = len(l)
    pos = 0
    num_steps = 0 
    
    while n > 3:
        num_steps += 1
        
        if n % 3 == 0: # n == 3k
            w = len(l) // 3
            s1 = l[:w]
            s2 = l[w:2 * w]
            s3 = l[2 * w: 3 * w]

            if sum(s1) < sum(s2):
                l = s1
                n = len(l)
            elif sum(s1) > sum(s2):
                l = s2
                n = len(l)
                pos += w
            else: # sum(s1) == sum(s2)
                l = s3
                n = len(l)
                pos += 2 * w

        else: # n == 3k + 1 or 3k + 2
            w = len(l) // 3
            s1 = l[:w]
            s2 = l[w:2 * w]
            s3 = l[2 * w: 3 * w]
            s4 = l[3 * w:]

            if sum(s1) < sum(s2):
                l = s1
                n = len(l)
            elif sum(s1) > sum(s2):
                l = s2
                n = len(l)
                pos += w
            else: # sum(s1) == sum(s2)
                num_steps += 1
                if sum(s1) > sum(s3):
                    l = s3
                    n = len(l)
                    pos += 2 * w
                elif sum(s1) == sum(s3):
                    l = s4
                    n = len(l)
                    pos += 3 * w
                
    num_steps += 1
    a[1] = num_steps
    if n == 1:
        a[0] = pos
    elif n == 2:
        num_steps += 1
        if l[0] < l[1]:
            a[0] = pos
        else:
            a[0] = pos + 1
    else: # n == 3
        num_steps += 1
        if l[0] < l[1]:
            a[0] = pos
        elif l[0] > l[1]:
            a[0] = pos + 1
        else:
            a[0] = pos + 2
    
    # adjust numsteps with find_fake
    # if the algorithm takes (n) routines, numsteps should be (n - 1)
    a[1] -= 1 
    
    return a
```

```python
N = 7
F = 1
G = 2
l = []
for i in range(N):
    l.append(G)
for i in range(N):
    l[i] = F
    print(l)
    a = find_fake(l,F)
    print("fake coin is in pos ",a[0], "Found after",a[1], "numsteps")
    b = find_fake_smart(l,F)
    print("fake coin is in pos ",b[0], "Found after",b[1], "numsteps")
    assert(a[0] == b[0])
    l[i] = G
    print(l)
```

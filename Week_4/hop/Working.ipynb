```python
def _hop_easy(a:'list',f:'int') -> 'int':
## n = len(a)  ##YOU CANNOT CALL len
  t = f
  h = 0
  while (True):
    if (a[t] == f):
      return h 
    else:
      t = a[t]
      h = h + 1
  return h
```

```python
def _hop(a:'list',f:'int') -> 'int':
    if a[f] == f:
        print('if section', a, f)
        return 0 
    else:
        nf = a[f]
        a[f] = a[nf]
        print('else section', a, f, nf)
        l = 1 + _hop(a, f)
        a[f] = nf
        return l
```
